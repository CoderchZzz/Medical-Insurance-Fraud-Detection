数据探索

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# 假设已经加载了数据集df
df = pd.read_csv('cleaned_data-0.csv')

# 字体
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

# 2.1 描述性统计
description = df.describe()
print(description)

# 2.2 数据可视化
## 设置可视化样式
sns.set(style="whitegrid")

## 绘制直方图
df.hist(figsize=(14, 14), bins=20)
plt.suptitle('各数值型字段的直方图')
plt.show()

## 绘制箱形图
# 确定有多少数值型字段需要绘制箱形图
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
n_cols = len(num_cols)

# 设置子图布局
n_rows = int(n_cols**0.5)
n_rows += n_rows * n_rows < n_cols  # 如果n_rows的平方小于n_cols，增加一行

# 创建一个大的图和多个子图
fig, axes = plt.subplots(n_rows, n_rows, figsize=(14, 14))
fig.suptitle('各数值型字段的箱形图')

# 在子图上绘制箱形图
for i, col in enumerate(num_cols):
    row = i // n_rows
    col_num = i % n_rows
    sns.boxplot(x=df[col], ax=axes[row, col_num])

# 隐藏空的子图
for i in range(n_cols, n_rows*n_rows):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()  # 调整子图的间距
plt.show()

## 使用散点图矩阵探索特征之间的关系
# 由于散点图矩阵可能非常密集，通常我们只选取一部分特征进行展示
sns.pairplot(df.select_dtypes(include=['float64', 'int64']).iloc[:, :5])
plt.suptitle('特征之间的散点图矩阵')
plt.show()
```
